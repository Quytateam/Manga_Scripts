```python
import pandas as pd
```

```python
mangas_df = pd.read_csv('../Collaborative Filtering/mangas1.csv')
genres_df = pd.read_csv(f'../Collaborative Filtering/genres.csv')
```

```python
mangas_df
```

```python
genres_df
```

```python
all_genres = genres_df["nameOnUrl"].tolist()
```

```python
records = []

# Điền dữ liệu vào danh sách
for index, row in mangas_df.iterrows():
    manga_genres = row["genre"].split("|")
    manga_dict = {"mangaId": row["_id"]}
    for genre in all_genres:
        manga_dict[genre] = 1 if genre in manga_genres else 0
    records.append(manga_dict)

# Tạo DataFrame từ danh sách các từ điển
result_df = pd.DataFrame.from_records(records)

# Hiển thị kết quả
print(result_df)
```

```python
result_df = result_df.set_index('mangaId')
print(result_df)
```

```python
result_df1 = result_df
```
